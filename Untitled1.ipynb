```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
faceImage = cv2.imread('face.png')
plt.imshow(faceImage[:,:,::-1]);plt.title("Face")
```

```python
faceImage.shape
```

```python
glassPNG = cv2.imread('glass.png',-1)
plt.imshow(glassPNG[:,:,::-1]);plt.title("glassPNG")
```

```python
glassPNG = cv2.resize(glassPNG,(190,50))
print("image Dimension ={}".format(glassPNG.shape))
```

```python
glassBGR = glassPNG[:,:,0:3]
glassMask1 = glassPNG[:,:,3]
```

```python
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(glassBGR[:,:,::-1]);plt.title('Sunglass Color channels');
plt.subplot(122);plt.imshow(glassMask1,cmap='gray');plt.title('Sunglass Alpha channel');
```

```python
faceWithGlassesNaive = faceImage.copy()
y1, y2 = 300, 350
x1, x2 = 220, 520

# Resize glasses to fit the selected area
glass_resized = cv2.resize(glassBGR, (x2 - x1, y2 - y1))

# Paste it
faceWithGlassesNaive[y1:y2, x1:x2] = glass_resized

plt.imshow(faceWithGlassesNaive[..., ::-1])
plt.axis("on")
```

```python
glassMask = cv2.merge((glassMask1, glassMask1, glassMask1))
glassMask = np.uint8(glassMask / 255)

faceWithGlassesArithmetic = faceImage.copy()

y1, y2 = 300, 350
x1, x2 = 220, 520

eyeROI = faceWithGlassesArithmetic[y1:y2, x1:x2]

glass_resized = cv2.resize(glassBGR, (eyeROI.shape[1], eyeROI.shape[0]))
mask_resized = cv2.resize(glassMask, (eyeROI.shape[1], eyeROI.shape[0]))

maskedEye = cv2.multiply(eyeROI, 1 - mask_resized)
maskedGlass = cv2.multiply(glass_resized, mask_resized)

eyeRoiFinal = cv2.add(maskedEye, maskedGlass)

faceWithGlassesArithmetic[y1:y2, x1:x2] = eyeRoiFinal

plt.figure(figsize=(20, 6))
plt.subplot(131); plt.imshow(maskedEye[..., ::-1]); plt.title("Masked Eye Region")
plt.subplot(132); plt.imshow(maskedGlass[..., ::-1]); plt.title("Masked Sunglass Region")
plt.subplot(133); plt.imshow(eyeRoiFinal[..., ::-1]); plt.title("Augmented Eye and Sunglass")
plt.axis("off")
```

```python
faceWithGlassesArithmetic[y1:y2, x1:x2] = eyeRoiFinal
plt.figure(figsize=[20,20]);
plt.subplot(121);plt.imshow(faceImage[:,:,::-1]); plt.title("Original Image");
plt.subplot(122);plt.imshow(faceWithGlassesArithmetic[:,:,::-1]);plt.title("With Sunglasses");
```
